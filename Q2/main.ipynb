```python
import pandas as pd
import PyPDF2
import tabula
import csv
```

```python
def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, "rb") as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        for page_number in range(num_pages):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()
    return text

def save_text_to_file(text, output_file):
    with open(output_file, "w", encoding="utf-8") as file:
        file.write(text)


pdf_file = "C:/Users/PC/Desktop/Results/Q2/Q2.pdf"

# to extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_file)

# Specify the output text file
output_file = "output.txt"

# Save the extracted text to a file
save_text_to_file(pdf_text, output_file)

print("Text saved to", output_file)
```

```python
column_headers = ['Description', 'Possible Root cause', 'Page Number']

file_path = 'final_storing.csv'

# writing column headers
with open(file_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(column_headers)
```

```python
# Function to extract description text from text file
def extract_descriptions(input_file):
    descriptions = []
    with open(input_file, 'r', encoding='utf-8') as file:
        is_description = False
        description_words = []

        for line in file:
            if 'Description' in line:
                is_description = True
                start_index = line.index('Description') + len('Description')
                description_words.extend(line[start_index:].split())
            elif 'Possible' in line:
                if is_description:
                    # If 'Possible' is found and we were in a description, append the description
                    description = ' '.join(description_words).strip()
                    if description:
                        descriptions.append(description)
                    # Reset for the next description
                    is_description = False
                    description_words = []
            elif is_description:
                description_words.extend(line.split())

        # Check for the last description in case 'Possible' is not encountered
        if is_description and description_words:
            description = ' '.join(description_words).strip()
            if description:
                descriptions.append(description)

    return descriptions


```

```python
# Function to extract possible root causes from text file
def extract_root_causes(input_file):
    root_causes = []
    with open(input_file, 'r', encoding='utf-8') as file:
        is_root_cause = False
        root_cause_words = []

        for line in file:
            if 'cause' in line:
                is_root_cause = True
                start_index = line.index('cause') + len('cause')
                root_cause_words.extend(line[start_index:].split())
            elif 'Troubleshooting' in line:
                if is_root_cause:
                    # If 'Troubleshooting' is found and we were in a root cause, append the root cause
                    root_cause = ' '.join(root_cause_words).strip()
                    if root_cause:
                        root_causes.append(root_cause)
                    # Reset for the next root cause
                    is_root_cause = False
                    root_cause_words = []
            elif is_root_cause:
                root_cause_words.extend(line.split())

        # Check for the last root cause in case 'Troubleshooting' is not encountered
        if is_root_cause and root_cause_words:
            root_cause = ' '.join(root_cause_words).strip()
            if root_cause:
                root_causes.append(root_cause)

    return root_causes



```

```python
# Function to extract page numbers 
def extract_data(input_file):
    data = []
    with open(input_file, 'r', encoding='utf-8') as file:
        is_diagnosis = False
        extracted_data = ""

        for line in file:
            if 'Diagnosis' in line:
                is_diagnosis = True
            elif 'Description' in line:
                if is_diagnosis:
                    # If data has been extracted after "Diagnosis" and before "Description"
                    data.append(extracted_data.strip())
                    is_diagnosis = False
                    extracted_data = ""  # Reset for the next data
            elif is_diagnosis:
                # Extracting the first data after "Diagnosis"
                extracted_data += line.strip() + " "

    return data


```

```python

# Extract descriptions, root causes, and number from 'output.txt'
descriptions = extract_descriptions('output.txt')
root_causes = extract_root_causes('output.txt')
extracted_page = extract_data('output.txt')

# Writing extracted descriptions, root causes, and number to 'final_storing.csv' file
with open('final_storing.csv', 'a', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    
    # Write each row containing description, root cause, and data
    for desc, cause, data in zip(descriptions, root_causes, extracted_page):
        writer.writerow([desc, cause, data])
```
